```python
import requests
from bs4 import BeautifulSoup

# This is a practice website specifically for scraping
url = "http://books.toscrape.com/"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
}

print("=== Book Store Product Ads Scraper ===\n")

try:
    response = requests.get(url, headers=headers, timeout=10)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all book products
        books = soup.find_all("article", class_="product_pod")

        print(f"Found {len(books)} products\n")

        for i, book in enumerate(books[:5], start=1):
            # Extract title
            title = book.h3.a['title']

            # Extract price
            price = book.find("p", class_="price_color").text

            # Extract rating
            rating = book.find("p", class_="star-rating")['class'][1]

            # Extract availability
            availability = book.find("p", class_="instock availability").text.strip()

            # Extract image URL
            image_url = "http://books.toscrape.com/" + book.find("img")['src'].replace('../', '')

            print(f"PRODUCT AD #{i}")
            print("=" * 50)
            print(f"Title       : {title}")
            print(f"Price       : {price}")
            print(f"Rating      : {rating} stars")
            print(f"Availability: {availability}")
            print(f"Image URL   : {image_url[:60]}..." if len(image_url) > 60 else f"Image URL   : {image_url}")
            print("-" * 50 + "\n")

    else:
        print(f"Failed. Status: {response.status_code}")

except Exception as e:
    print(f"Error: {e}")
```
